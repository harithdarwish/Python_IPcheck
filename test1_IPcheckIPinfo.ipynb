```python
import requests

API_KEY = #YOUR_API_KEY

def test_lite_api():
    # Lite endpoint: api.ipinfo.io/lite/
    url = 'https://api.ipinfo.io/lite/8.8.8.8'
    headers = {'Authorization': f'Bearer {API_KEY}'}
    
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        print("✅ IPinfo Lite Success!")
        return True
    else:
        print(f"❌ Failed: {response.status_code}")
        return False

api_works = test_lite_api()
```

```python
import requests
import pprint

# --- CONFIGURATION ---
API_KEY = #YOUR_API_KEY
IP_TO_CHECK = '8.8.8.8'

# 1. Use the /lite/ endpoint for unlimited requests
url = f'https://api.ipinfo.io/lite/{IP_TO_CHECK}'

# 2. IPinfo uses "Authorization: Bearer" instead of "Key"
headers = {
    'Accept': 'application/json',
    'Authorization': f'Bearer {API_KEY}'
}

# 3. Request
response = requests.get(url, headers=headers)

# 4. Print results
if response.status_code == 200:
    print("--- RAW JSON DATA ---")
    pprint.pprint(response.json())
else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
import sys
MASTER_DATA = []

if api_works:
    # Check the first 900 for now
    test_list = my_ips[:1000] 
    
    for index, ip in enumerate(test_list):
        # Progress Counter
        sys.stdout.write(f"\r🔍 Checking: {index + 1} of {len(test_list)}... ({ip})")
        sys.stdout.flush()

        url = f'https://api.ipinfo.io/lite/{ip}'
        headers = {'Authorization': f'Bearer {API_KEY}'}
        
        try:
            response = requests.get(url, headers=headers)
            # The Lite JSON has: 'ip', 'as_name', 'country',
            MASTER_DATA.append(response.json())
        except:
            continue
```

```python
import pandas as pd

if MASTER_DATA:
    df = pd.DataFrame(MASTER_DATA)

    # LITE SPECIFIC COLUMNS
    columns_to_keep = {
        'ip': 'IP Address',
        #'country': 'Country Name',
        'continent': 'Continent',
        #'asn': 'ASN Number',
        'as_name': 'Organization/ISP',
        #'as_domain': 'Domain',
    
    }
    
    # Filter only what exists in the data
    existing = [c for c in columns_to_keep.keys() if c in df.columns]
    df_final = df[existing].rename(columns=columns_to_keep)

    df_final.to_excel("IPinfo_Lite_Report.xlsx", index=False)
    print("\n✅ Excel Created: IPinfo_Report.xlsx")
```
